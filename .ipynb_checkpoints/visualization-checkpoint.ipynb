```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsilon = [1,2,4,8]
scale1 = [0.89, 0.85, 0.45, 0.35]
scale2 = [0.38, 0.18, 0.083, 0.049]

```

```python
plt.plot(epsilon,  scale2, marker = 'o',linestyle = 'dashdot',label="500Mb")
plt.plot(epsilon, scale1, marker = 'o',linestyle = 'dashdot',label="100Mb")
plt.xlabel("Privacy Budget")
plt.ylabel("Error Level")
plt.legend()
plt.savefig("./result.png")
```
